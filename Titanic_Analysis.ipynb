# Titanic Survivor Analysis

## Summary
This notebook summarizes the data analysis process used in the analysis of the Titanic passenger data. The goal of the analysis was to understand the factors that made people more likely to survive.  The dataset for this analysis is a sample of made up of data for 891 of the 2,224 passengers on board the Titanic. For simplicity sake, this analysis refers to all the people in the sample as passengers.  

## Findings
Of the 891 passengers in the sample analyzed, 342 passengers survived or 38 percent of passengers. 

The most important determinants of a passenger's survival were gender, class, and age. Approximately 74 percent of female passengers survived, compared to only 19 percent of men. Women and children (passengers 15 years or younger) travelling in first and second class had the highest survival rate of any demographic group.  Women travelling in first class and second class had a survival rate of 97 and 92 percent respectively; children traveling in first or second class had a survival rate of 96 percent. Female passengers traveling in third class had a survival rate of 50 percent, which was significantly lower than women traveling in first or second class, but still higher than the overall survival rate. 

For male passengers, the most important determinant of survival were class and whether they were traveling alone or in a group (as measured by the number of passengers on a ticket). Male passengers traveling in first class had a survival rate of 37 percent, about twice as high as the survival rate of male passengers in second or third class. The survival rate of male passengers traveling in a group was 28 percent, compared to only 15 percent for male passengers traveling on individual tickets. 

Overall, age was an important determinant of survival for passengers that were 25 years or younger. Children, which we defined as travelers 15 years or younger, had a survival rate of 59 percent. Children made up 11.6 percent of passengers but 16.9 percent of survivors. Class was a key determinant in the survival of children. Of the 25 children that were traveling in first or second class, only one perished. In contrast, 33 of the 58 children traveling in third class perished. Passenger between the ages of 16 and 25 made up a disproportionate share of those who perished. Although passengers in this group made up nearly 31 percent of passengers, they only made up 26 percent of survivors. 


### Limitations
Although this analysis can be used to understand the factors that contributed to a passenger's chances of survival, the sample used for the analysis accounts for about 40 percent of the actual passengers. Statistics for the actual population may vary. 


## Data Exploration and Clean-Up
Overall, the data is relatively clean and there was no need for major modifications. Pandas assigned all of the correct variable types for all of the variables. I did not change the passenger ID to a string because I did not use it to form any of my analysis. All names and ID numbers were unique. Of the variables analyzed, age was the only one the only one with missing data (177 records did not have data).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
dat = pd.read_csv("titanic-data.csv")

# Set column names to lower case for easier coding
dat.columns = map(str.lower, dat.columns)
# Print column names and types
colnames = list(dat)
for item in colnames:
    print(str(item + ": " + str(dat[item].dtype)))

# Check if the number of unique names is the same as the number of unique recods
if  dat.groupby("name").size().sum() == dat.groupby("passengerid").size().sum():
    recs = dat.groupby("name").size().sum()
    print("\nThere are {} records. Each name and ID is unique.".format(recs))

print("\nNumber of values in column")
print(dat.count())
```

## Exploratory Analysis

### Overall Survival Rate

```python
# Total number of records. Function below counts unique passenger IDs and then adds them up
by_survived = dat.groupby("survived").size()
survivors = by_survived[1]
tot_recs = by_survived.sum()

# share of survivors
per_survivors = round(survivors / float(tot_recs), 2)
print("Of the {} passengers, {} survived, or aproximately {}% the passengers.".format(tot_recs, survivors, per_survivors*100))
```

### Survival by Gender

```python
# Survival rate by gender
gender_analysis = pd.crosstab(dat.sex, dat.survived)
gender_analysis.columns = ["perished", "survived"]
gender_analysis["count_by_gender"] = gender_analysis["perished"] + gender_analysis["survived"]
gender_analysis["survival_rate"] = gender_analysis["survived"]/gender_analysis["count_by_gender"]

# Share of passengers and survivors by gender
gender_analysis["share_of_survivors"] = gender_analysis["survived"]/gender_analysis["survived"].sum()
gender_analysis["share_of_passengers"] = gender_analysis["count_by_gender"]/gender_analysis["count_by_gender"].sum()

# Function for changing the tables
gender_analysis
```

```python
# Graph the change in distribution by gender among passengers and survivors

# This function is used to format the results to show up as a percentages
def as_percent(x):
    return("{:.3}".format(x*100))

l1 = gender_analysis["share_of_passengers"].tolist()
l2 = gender_analysis["share_of_survivors"].tolist()
l3 = ["female", "male"]
l4 = ["share of passengers", "share of survivors"]

plt.figure(figsize=(7,5))
plt.plot([l1, l2])
plt.axis([-0.5,1.65,.20,.80])
for n, gender in zip(l1, l3):
    plt.annotate((gender + ": "+ as_percent(n) + "%"), xy = (-0.41, n), xytext=(-0.41, (n)))
for n in l2:
    plt.annotate((as_percent(n) + "%"), xy = (1.01, n), xytext=(1.01, (n)))

plt.annotate("share of passengers", xy = (-0.41, 0.73), xytext = (-0.41, 0.75))
plt.annotate("share of survivors", xy = (1.01, 0.73), xytext = (1.01, 0.75))
#for n, age_group in zip(l2, l3):
#    plt.annotate((str(n) + "%; "+ str(age_group)), xy = (1, n), xytext=(1.03, ytext))

plt.xticks([])
plt.yticks([])
plt.show()
```

### Survival by Age

After plotting the data on a histogram, I decided to group passengers by six bins. I wanted the first bin to only include minors. I assumed that there was a good likelihood that some of the passengers traveling alone were under the age of 18, so I analyzed the data of passengers between the ages of 11 and 17 traveling without a parent and determined that the best most appropriate cut off age for my "minors" group was 15. 

**Guiding questions**
* What was youngest age that passengers could travel alone?
* What was the survival rate of passengers by age group?
* What was the share of passengers and survivals by age group?

```python
#check the minimum and max ages in the data
print("Min age was: " + str(dat["age"].min()))
print("Max age was: " + str(dat["age"].max()))
print("Records with missing age data: " + str(dat["age"].isnull().sum()))

# check the age at which travelers could travel alone. I first printed to see which ages had the most counts. 
minors_noparents = dat[(dat["age"] < 18) & (dat["age"] >= 10) & (dat["parch"] == 0)]
minors_noparents = pd.Series(minors_noparents.groupby("age").size())
all_minors = dat[(dat["age"] < 18) & (dat["age"] > 10)]
all_minors = pd.Series(all_minors.groupby("age").size())

comparison = pd.DataFrame({"minors_count":all_minors, "minors_noparents":minors_noparents})
print(comparison)
```

```python
# Survival by age group

# Add another column with age group in order simplify coding 
agebins = [0, 16, 26, 36, 46, 56, 81]
agegroups = ["0 to 15", "16 to 25", "26 to 35", "36 to 45", "46 to 55", "56 to 80"]
dat["age_group"] = pd.cut(dat.age, bins = agebins, labels = agegroups, right = False, include_lowest = True)

# Survival rate by age group
age_analysis = pd.crosstab(dat.age_group, dat.survived)
age_analysis.columns = ["perished", "survived"]
age_analysis["passengers_by_agegroup"] = (age_analysis["perished"]+age_analysis["survived"])
age_analysis["survival_rate"] = age_analysis["survived"]/age_analysis["passengers_by_agegroup"]

# Share of passenger and survivals by age group. 

## Get the total number of passengers and survivor that have age data
passengers_w_agedata = len(dat["age"].dropna())
survivors_w_agedata = len(dat[(dat["age"].isnull() == False) & dat["survived"] == 1])
## Calculate share of passengers and surviors by age group
age_analysis["share_of_passengers"] = age_analysis["passengers_by_agegroup"]/passengers_w_agedata
age_analysis["share_of_survivors"] = age_analysis["survived"]/(age_analysis["survived"].sum())
age_analysis["difference_in_share_of_passengers_survivors"] = age_analysis["share_of_survivors"] - age_analysis["share_of_passengers"]

age_analysis
```

### Party size analysis
**Guiding questions:**
* Did party size influence survival rate?
* Were male passengers traveling as part of a group more likely to survive?

```python
# Impact of traveling in a group

# The function below is used to quickly reformat results returned as floats.
def share(numerator, denominator):
    return round((numerator/float(denominator)), 3)*100

# Get survival data by ticket.
party_analysis = pd.crosstab(dat.ticket, dat.survived)
party_analysis["passengers_per_ticket"] = party_analysis[0]+party_analysis[1]

# Create a new column with the variable denoting if the ticket was for single or multiple people
party_analysis["party_type"] = pd.cut(party_analysis.passengers_per_ticket, bins = [0, 1, 8], labels = ["single", "multiple"],right = True)
party_analysis = party_analysis.reset_index()
party_analysis.columns =['ticket', 'perished', 'survived', 'passengers_per_ticket', 'party_type']
mulparty = party_analysis[["passengers_per_ticket", "survived"]][party_analysis["party_type"] == "multiple"].sum()
singleparty = party_analysis[["passengers_per_ticket", "survived"]][party_analysis["party_type"] == "single"].sum()
mulparty_survival_rate = share(mulparty[1], mulparty[0])
singleparty_survival_rate = share(singleparty[1], singleparty[0])

print("{}% of passengers travelling in a group survived, compared to {:.3}% of passengers traveling alone".format(mulparty_survival_rate, singleparty_survival_rate))
```

```python
# Analysis by gender and party size
party_analysis_by_gender = pd.crosstab(dat.ticket, [dat.sex, dat.survived]).reset_index()

# Concatenate the gender analysis table with the table that has party_type data. 
# Since rownames were the same, there was no need to manipulate the data. 
ticket_partytype = party_analysis[["ticket", "party_type"]]
party_analysis_by_gender = pd.concat([party_analysis_by_gender, ticket_partytype], axis=1)

# Calculations
party_analysis_by_gender.columns = ["ticket","female_perished","female_survived","male_perished","male_survived", "ticket", "party_type"]
party_analysis_by_gender["male_count"] = party_analysis_by_gender["male_perished"]+party_analysis_by_gender["male_survived"]
male_in_mulparty = party_analysis_by_gender[["male_survived", "male_count"]][party_analysis_by_gender["party_type"] == "multiple"].sum()
male_in_sinparty = party_analysis_by_gender[["male_survived", "male_count"]][party_analysis_by_gender["party_type"] == "single"].sum()

# Print results
print("{:.3}% of males in multi-party tickets survived".format(share(male_in_mulparty[0], male_in_mulparty[1])))
print("{:.3}% of males in single-party tickets survived".format(share(male_in_sinparty[0], male_in_sinparty[1])))
```
